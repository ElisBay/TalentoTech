```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

# Define la función a minimizar
def funcion(x):
  return x**2 + 2*x + 1
```

```python
# Define la derivada de la función
def derivada(x):
  return 2*x + 2
```

```python

# Define la tasa de aprendizaje
tasa_aprendizaje = 0.1
print(tasa_aprendizaje)

```

# Se solicita a Gemini generar codigo para calcular la tasa de aprendizaje

```python
# Función para calcular la tasa de aprendizaje
def calcular_tasa_aprendizaje(x_actual, direccion_descenso):
  """
  Calcula la tasa de aprendizaje usando una búsqueda de línea simple.

  Args:
    x_actual: El punto actual en el espacio de búsqueda.
    direccion_descenso: La dirección de descenso (negativo del gradiente).

  Returns:
    La tasa de aprendizaje óptima.
  """

  # Rango de valores para la tasa de aprendizaje
  tasas_aprendizaje = np.linspace(0.01, 1, 100)

  # Inicializa la mejor tasa de aprendizaje y el mejor valor de la función
  mejor_tasa_aprendizaje = 0
  mejor_valor_funcion = float('inf')  # Inicializa con infinito

  # Busca la mejor tasa de aprendizaje
  for tasa in tasas_aprendizaje:
    x_nuevo = x_actual + tasa * direccion_descenso
    valor_funcion = funcion(x_nuevo)

    if valor_funcion < mejor_valor_funcion:
      mejor_valor_funcion = valor_funcion
      mejor_tasa_aprendizaje = tasa

  return mejor_tasa_aprendizaje

# Ejemplo de uso
x_actual = 2  # Punto inicial
direccion_descenso = -derivada(x_actual)  # Dirección de descenso

tasa_aprendizaje_optima = calcular_tasa_aprendizaje(x_actual, direccion_descenso)

print("Tasa de aprendizaje óptima:", tasa_aprendizaje_optima)
```

```python
import numpy as np


# Define la función de descenso de gradiente
def descenso_gradiente(valor_inicial, precision, tasa_aprendizaje):
  x = valor_inicial
  while True:
    x_anterior = x
    x = x - tasa_aprendizaje * derivada(x)
    if abs(x - x_anterior) < precision:
      break
  return x

# Define los valores iniciales, la precisión y la tasa de aprendizaje
valores_iniciales = [0, 1, 2]
precision = 0.0001
tasas_aprendizaje = [0.1, 0.01, 0.001]

# Ejecuta el descenso de gradiente para cada combinación de valores
for valor_inicial in valores_iniciales:
  for tasa_aprendizaje in tasas_aprendizaje:
    minimo = descenso_gradiente(valor_inicial, precision, tasa_aprendizaje)
    print(f"Valor inicial: {valor_inicial}, Tasa de aprendizaje: {tasa_aprendizaje}, Mínimo: {minimo}")
```

```python
# Paso 1: Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Paso 2: Crear el dataframe con columnas x1, x2 y una columna 'y' como variable dependiente
np.random.seed(42)  # Para reproducibilidad
```

```python
# Generar datos aleatorios
x1 = np.random.rand(20) * 10  # 20 filas para x1
x2 = np.random.rand(20) * 10  # 20 filas para x2
y = 2 * x1 + 3 * x2 + np.random.randn(20)  # y es una combinación lineal de x1 y x2 con algo de ruido
```

```python
# Crear DataFrame
df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
df
```

```python
# Paso 3: Graficar x1 vs y, y x2 vs y
plt.figure(figsize=(14,6))
```

```python
# Gráfico de x1 vs y
plt.subplot(1, 2, 1)
plt.scatter(df['x1'], df['y'], color='blue')
plt.title('x1 vs y')
plt.xlabel('x1')
plt.ylabel('y')
```

```python
# Gráfico de x2 vs y
plt.subplot(1, 2, 2)
plt.scatter(df['x2'], df['y'], color='red')
plt.title('x2 vs y')
plt.xlabel('x2')
plt.ylabel('y')
plt.show()
```

```python
# Paso 4: Hallar coeficientes e interceptos con una regresión lineal
X = df[['x1', 'x2']]  # Variables independientes
y = df['y']  # Variable dependiente
```

```python
# Paso 4: Realizar regresión lineal múltiple
X = df[['x1', 'x2']]  # Variables independientes (x1 y x2)
y = df['y']  # Variable dependiente

# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo con los datos
model.fit(X, y)
```

```python
# Paso 5: Hallar los coeficientes e interceptos
coeficientes = model.coef_  # Coeficientes de x1 y x2
intercepto = model.intercept_  # Intercepto
```

```python
# Paso 6: Imprimir la ecuación de la regresión lineal
print(f"Ecuación de la regresión: y = {intercepto:.4f} + {coeficientes[0]:.4f}*x1 + {coeficientes[1]:.4f}*x2")
```

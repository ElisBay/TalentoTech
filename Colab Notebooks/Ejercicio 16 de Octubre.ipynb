```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# 1. Cargar el dataset MNIST
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

# 2. Normalizar los valores de los píxeles a un rango de 0 a 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# 3. Reshape para agregar la dimensión del canal (1 para imágenes en escala de grises)
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# 4. Definir el modelo de la red convolucional
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(10, activation='softmax'))

# 5. Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 6. Entrenar el modelo
model.fit(train_images, train_labels, epochs=5)

# 7. Evaluar el modelo
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nPrecisión en el conjunto de prueba:', test_acc)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import os

# 1. Define el directorio que contiene tus imágenes
# Reemplaza 'path/to/your/images/directory' con la ruta real a tu directorio de imágenes.
images_dir = 'path/to/your/images/directory'  # Cambia esto por la ruta correcta

# Verifica si el directorio existe
if not os.path.exists(images_dir):
    print(f"Error: El directorio '{images_dir}' no existe.")
else:
    # 2. Obtén una lista de todos los archivos de imagen en el directorio
    image_files = [f for f in os.listdir(images_dir) if os.path.isfile(os.path.join(images_dir, f))]

    # 3. Bucle para procesar cada archivo de imagen y hacer predicciones
    for image_file in image_files:
        img_path = os.path.join(images_dir, image_file)

        try:
            # Cargar y preprocesar la imagen
            img = image.load_img(img_path, target_size=(28, 28), color_mode="grayscale")
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = img_array / 255.0  # Normalización

            # Realizar la predicción
            prediction = model.predict(img_array)
            predicted_class = np.argmax(prediction[0])

            # Mostrar el resultado
            print(f"Imagen: {image_file}, Clase Predicha: {predicted_class}")

        except Exception as e:
            print(f"Error procesando la imagen {image_file}: {e}")
```

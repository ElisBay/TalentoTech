```python
from sklearn.metrics import confusion_matrix
```

```python
# Valores de la matriz de confusión
matriz_confusion = [[162, 4],
                    [11, 22]]
```

```python
# Extraer verdaderos positivos, falsos positivos, falsos negativos y verdaderos negativos
TN = matriz_confusion[0][0]
FP = matriz_confusion[0][1]
FN = matriz_confusion[1][0]
TP = matriz_confusion[1][1]
print(TN)
print(FP)
print(FN)
print(TP)
```

```python
# Cálculo del accuracy Positivo
total = TP + FP + FN + TN
accuracy = (TP+ TN) / total
accuracy
```

```python
# Precisión
precision = TP / (TP + FP)

# Mostrar el resultado de la precisión
print(f'Precisión: {precision}')
```

```python
Recall=TP/(FN + TP)
Recall
```

```python
#F1_Score=2*(precision * recall) / (precision + recall)
F1_Score = 2*(precision * Recall) / (precision + Recall)
F1_Score
```

```python
# Cálculo del accuracy Negativo
total = TP + FP + FN + TN
accuracy2 = (FP+ FN) / total
accuracy2
```

```python
# Precisión Negativa
precision2 = TN / (TN + FN)

# Mostrar el resultado de la precisión
print(f'Precisión2: {precision2}')
```

```python
Recall2=TN/(TN + FP)
Recall2
```

```python
#F1_Score2=2*(precision2 * Recall2) / (precision2 + Recall2)
F1_Score2 = 2*(precision2 * Recall2) / (precision2 + Recall2)
F1_Score
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score

# Datos de la matriz de confusión
conf_matrix = np.array([[162, 4],
                        [11, 22]])
```

```python
# Crear probabilidades simuladas para la clase positiva (1)
y_scores = np.array([0.9] * conf_matrix[0, 0] +  # TP
                    [0.8] * conf_matrix[0, 1] +  # FP
                    [0.1] * conf_matrix[1, 0] +  # FN
                    [0.7] * conf_matrix[1, 1])    # TN
```

```python
# Definir etiquetas verdaderas y probabilidades predichas
y_true = [0] * TN + [1] * TP + [0] * FP + [1] * FN  # Etiquetas verdaderas
y_prob = [0.1] * TN + [0.9] * TP + [0.5] * FP + [0.2] * FN  # Probabilidades predichas
```

```python
# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_true, y_prob)
```

```python
# Calcular el AUC
auc = roc_auc_score(y_true, y_prob)
auc
```

```python
# Graficar la curva ROC
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Línea de referencia
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()

# Mostrar AUC
auc
```

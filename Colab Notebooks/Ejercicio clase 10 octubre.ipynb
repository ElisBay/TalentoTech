```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Cargar el dataset Iris
iris = load_iris()
X = iris.data
iris.data
```

```python
X.shape
```

```python
# Lista para almacenar las inercias
inertias = []
```

```python
# Probar K-means para diferentes valores de k (número de clústeres)
K = range(1, 11)  # Probar con k desde 1 hasta 10
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)  # Guardar la inercia para cada k
    print(f"K={k}, Inercia: {inertias[-1]}")

```

```python
# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(K, inertias, 'bo-', markersize=8)
plt.xlabel('Número de clústeres (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para encontrar el número óptimo de clústeres')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
len(kmeans.labels_)
```

```python
#predice los cluster
y_kmeans = kmeans.predict(X)
```

```python
# Import the pandas library and assign it to the alias 'pd'
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
# Import the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()

#visualice kmeans clustering Vs Actual labels
#convert to dataFrame foer easir plotting
iris_df= pd.DataFrame(iris.data, columns=iris.feature_names)
# Access the target variable from the iris dataset and assign it to 'y'
y = iris.target
iris_df['Actual Species'] = y
iris_df['cluster'] = y_kmeans
```

```python
# Convertir el array X a un DataFrame de pandas
iris_df = pd.DataFrame(X, columns=iris.feature_names)
```

```python
# Aplicar KMeans y agregar los clusters al DataFrame
kmeans = KMeans(n_clusters=3, random_state=42)  # Usamos 3 clusters como ejemplo
iris_df['cluster'] = kmeans.fit_predict(X)  # Guardamos los clusters predichos
```

```python
#visualize kmeans clustering Vs Actual labels
#convert to dataFrame for easier plotting
sns.pairplot(iris_df, hue='cluster', palette='coolwarm', diag_kind='kde')
plt.suptitle('Kmeans Clustering of Iris Data', y=1.02)
plt.show()
```

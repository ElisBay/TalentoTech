```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(6, 6))

# Configuración del fondo en color verde fluorescente
fig.patch.set_facecolor('#39FF14')  # Verde fluorescente

# Cuadrado 1 (arriba izquierda) - Magenta
rect1 = patches.Rectangle((0, 0.5), 2, 2, linewidth=0, edgecolor='none', facecolor='#FF00FF')
ax.add_patch(rect1)

# Cuadrado 2 (arriba derecha) - Azul Rey
rect2 = patches.Rectangle((0.5, 0.5), 0.5, 0.5, linewidth=0, edgecolor='none', facecolor='#0000FF')
ax.add_patch(rect2)

# Cuadrado 3 (abajo izquierda) - Magenta
rect3 = patches.Rectangle((0, 0), 0.5, 0.5, linewidth=0, edgecolor='none', facecolor='#FF00FF')
ax.add_patch(rect3)

# Cuadrado 4 (abajo derecha) - Azul Rey
rect4 = patches.Rectangle((0.5, 0), 0.5, 0.5, linewidth=0, edgecolor='none', facecolor='#0000FF')
ax.add_patch(rect4)

# Dibujar la cruz en el centro de color amarillo
# Línea vertical
plt.plot([0.5, 0.5], [0, 1], color='yellow', linewidth=15)
# Línea horizontal
plt.plot([0, 1], [0.5, 0.5], color='yellow', linewidth=15)

# Configurar los límites y quitar ejes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')  # Desactivar los ejes para una visualización limpia

plt.show()
```

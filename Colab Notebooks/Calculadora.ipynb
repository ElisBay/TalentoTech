```python
# Funciones matemáticas
def operar(a, b, operacion):
    if operacion == '1':
        return a + b
    elif operacion == '2':
        return a - b
    elif operacion == '3':
        return a * b
    elif operacion == '4':
        return "Error: No se puede dividir entre cero." if b == 0 else a / b

# Menú de la calculadora
def calculadora():
    print("Bienvenido a la calculadora!")

    # Validar que el operador sea correcto
    opciones_validas = ['1', '2', '3', '4']
    while True:
        print("Elige una opción:")
        print("1. Sumar")
        print("2. Restar")
        print("3. Multiplicar")
        print("4. Dividir")

        # Recibir la opción del usuario
        opcion = input("Introduce una opción (1, 2, 3, 4): ")

        if opcion in opciones_validas:
            break  # Si el operador es válido, salir del bucle
        else:
            print("Operador no válido. Por favor, introduce una opción correcta (1, 2, 3, 4).")

    # Recibir los números a operar
    num1 = float(input("Introduce el primer número: "))
    num2 = float(input("Introduce el segundo número: "))

    # Realizar la operación
    resultado = operar(num1, num2, opcion)
    print(f"El resultado es: {resultado}")

# Ejecutar la calculadora
calculadora()
```

```python
import random

def mostrar_ahorcado(intentos):
    estados = [
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        =========
        """,  # 6 intentos
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        =========
        """,  # 5 intentos
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        =========
        """,  # 4 intentos
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        =========
        """,  # 3 intentos
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        =========
        """,  # 2 intentos
        """
           -----
           |   |
           O   |
               |
               |
               |
        =========
        """,  # 1 intento
        """
           -----
           |   |
               |
               |
               |
               |
        =========
        """  # 0 intentos
    ]
    print(estados[intentos])

def adivina_la_palabra():
    print("¡Bienvenido al juego del Ahorcado!")

    # Lista de palabras
    palabras = ['python', 'programacion', 'tecnologia', 'desarrollo', 'inteligencia']

    # Seleccionar una palabra aleatoria
    palabra_secreta = random.choice(palabras)

    # Crear una lista de guiones bajos para mostrar el progreso del jugador
    progreso = ['_'] * len(palabra_secreta)

    # Contador de intentos
    intentos = 6
    letras_acertadas = set()  # Usar un conjunto para evitar duplicados
    letras_falladas = set()

    print(f"La palabra tiene {len(palabra_secreta)} letras. ¡Tienes {intentos} intentos para adivinarla!")

    while intentos > 0:
        mostrar_ahorcado(intentos)
        print(" ".join(progreso))

        entrada_usuario = input("Adivina una letra o la palabra completa: ").lower()

        if len(entrada_usuario) > 1:  # Intento de adivinar la palabra completa
            if entrada_usuario == palabra_secreta:
                print(f"¡Felicidades! Adivinaste la palabra: {palabra_secreta}")
                break
            else:
                intentos -= 3
                print(f"Palabra incorrecta. Te quedan {intentos} intentos.")
                continue
        elif entrada_usuario in letras_acertadas or entrada_usuario in letras_falladas:
            print(f"Ya adivinaste o fallaste la letra '{entrada_usuario}', intenta con otra.")
            continue

        # Comprobar si la letra está en la palabra secreta
        if entrada_usuario in palabra_secreta:
            letras_acertadas.add(entrada_usuario)
            for i, char in enumerate(palabra_secreta):
                if char == entrada_usuario:
                    progreso[i] = entrada_usuario
            print(f"¡Bien! La letra '{entrada_usuario}' está en la palabra.")
        else:
            letras_falladas.add(entrada_usuario)
            intentos -= 1
            print(f"Lo siento, la letra '{entrada_usuario}' no está en la palabra. Te quedan {intentos} intentos.")

        # Verificar si se ha adivinado toda la palabra
        if '_' not in progreso:
            print(f"¡Felicidades! Adivinaste la palabra: {palabra_secreta}")
            break
    else:
        mostrar_ahorcado(intentos)
        print(f"¡Demasiados intentos! El ahorcado ha perdido la batalla. La palabra era: {palabra_secreta}")

# Ejecutar el juego
adivina_la_palabra()
KeyboardInterrupt
```

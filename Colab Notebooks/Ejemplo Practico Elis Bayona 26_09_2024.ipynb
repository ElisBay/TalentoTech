```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create the dataset
# Let´s build the values lists
value_n_list = ['Ana', 'Frank', 'Jorge', 'Diana', 'John', 'Andres', 'Juan', 'Diego', 'Alan', 'Jan']  # Added quotes around names to denote them as strings
value_N1_list = [5, 7, 8, 9, 6, 8, 8, 9, 8, 7]  # [Nota 1]
value_N2_list = [8, 9, 5, 7, 8, 9, 10, 8, 6, 3]  # [Nota 2]

# value_Final_list will be calculated: Final= average(N1+N2)/2
# Let´s use a dictionary ({key;value}) with the format {"str":list}
data_dict = {
    "Name": value_n_list,
    "N1": value_N1_list,
    "N2": value_N2_list
}

# Print the dictionary
print(data_dict)  # Changed Print to print

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

# Calculate the average and add the 'Final' column
df['Final'] = df[['N1', 'N2']].mean(axis=1)  # Final is the average of N1 and N2

# Display the DataFrame

df
```

# **Distancia Jaccard**

```python
# Definimos dos conjuntos A y B
A = {1, 2, 3}
B = {2, 3, 4, 5}

# Paso 1: Encontrar la intersección A ∩ B
# La intersección es el conjunto de los elementos que están en ambos conjuntos
interseccion = A & B  # Esto nos da {2, 3}

# Paso 2: Encontrar la unión A ∪ B
# La unión es el conjunto de todos los elementos que están en A, en B, o en ambos
union = A | B  # Esto nos da {1, 2, 3, 4, 5}

# Paso 3: Aplicar la fórmula de la distancia de Jaccard
# Distancia de Jaccard(A, B) = 1 - (tamaño de la intersección / tamaño de la unión)
distancia_jaccard = 1 - len(interseccion) / len(union)  # 1 - 2/5 = 0.6

# Mostramos los resultados
print("Intersección:", interseccion)
print("Unión:", union)
print("Distancia de Jaccard:", distancia_jaccard)
```

Distancia Hamming

```python
# Ejemplo 2: Cadenas de texto
cadena1_texto = 'karolin'
cadena2_texto = 'kathrin'

# Inicializamos otra variable para contar las diferencias en las cadenas de texto
distancia_hamming_texto = 0

# Recorremos ambas cadenas al mismo tiempo, comparando carácter por carácter
for char1, char2 in zip(cadena1_texto, cadena2_texto):
    if char1 != char2:  # Si los caracteres en la misma posición son diferentes
        distancia_hamming_texto += 1  # Aumentamos el contador de diferencias

```

```python
# Mostramos el resultado de la distancia de Hamming para las cadenas de texto
print("Distancia de Hamming (texto):", distancia_hamming_texto)
```
